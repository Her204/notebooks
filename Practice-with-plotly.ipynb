```
df= pd.read_csv("fao_data_forest_data.csv")

df = df.loc[(df.ELEM==df.ELEM.unique()[1])&(df.GOAL==df.GOAL.unique()[1])]
df = df.sort_values("VALUE",ascending=False)

pivot = df.pivot(columns="YEAR",index="AREA-NEW",values="VALUE")
pivot["TOTAL"] = pivot.sum(axis=1)
pivot = pivot.sort_values("TOTAL",ascending=False)
pivot = pivot.iloc[:6]
pivot
```

```
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
url = "https://upload.wikimedia.org/wikipedia/commons/thumb/6/60/Cat_silhouette.svg/20px-Cat_silhouette.svg.png"
url_2 = "https://upload.wikimedia.org/wikipedia/commons/thumb/a/a9/Michis.png/220px-Michis.png"
def rgb2gray(rgb):
    return np.dot(rgb, [0.299, 0.587, 0.144])
img = plt.imread(url_2)
gray = rgb2gray(img)
gray
```

```

from scipy import ndimage
gray = rgb2gray(img)
print(gray)
data = go.Heatmap(z=gray/255,colorscale="gray")

fig = go.Figure(data=data)

fig.show()
```
